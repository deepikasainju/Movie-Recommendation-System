```python
import pandas as pd
df = pd.read_csv('/content/movies.csv')
df
```

```python
df['genres'].apply(lambda x: x.replace("|"," "))
```

```python
feature =  df.title + " "+ df['genres'].apply(lambda x: x.replace("|"," "))
feature = feature.str.lower()
feature
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(feature.values)
```

```python
df.dtypes
```

```python
ids = df.movieId.astype(str)
metadata = df.to_dict('records')
```

```python
import chromadb
client = chromadb.Client() # client banaune tespaxi collection banaune tespaxi document upsert or insert garne
```

```python
collection = client.create_collection("vector_db")
```

```python
collection.add(ids = ids.tolist(),
               documents = feature.values.tolist(),
               embeddings = embeddings,
               metadatas =  metadata)
```

```python
user_input= input("What type of movies do you like? ")
```

```python
recommendation = collection.query(query_texts = [user_input], n_results = 4)
```
